```python
from requests import get
from bs4 import BeautifulSoup
from time import sleep
from random import randint
import numpy as np
import pandas as pd
```

```python
urls=["https://www.imdb.com/search/title/?groups=bottom_1000&count=250&ref_=adv_prv",
"https://www.imdb.com/search/title/?groups=bottom_1000&count=250&start=251&ref_=adv_nxt",
"https://www.imdb.com/search/title/?groups=bottom_1000&count=250&start=501&ref_=adv_nxt",
"https://www.imdb.com/search/title/?groups=bottom_1000&count=250&start=751&ref_=adv_nxt"]

```

```python
Name=[]
Rating=[]
MetaScore=[]
Votes=[]
Genre=[]
Year=[]
Gross=[]
```

```python
for url in urls:
    print("DOing URL"+str(url))
    code=get(url)
    soup=BeautifulSoup(code.text, "html.parser")
    div_movies=soup.find_all('div', class_="lister-item mode-advanced")
    for i in range(len(div_movies)):
        sample=div_movies[i]
        #getting name
        Name.append(sample.h3.a.text)
        temp=sample.find('span', class_="lister-item-year text-muted unbold").text
        Year.append(int(temp[-5:-1]))
        x=sample.find('div', class_="inline-block ratings-metascore")
        if(x):
                MetaScore.append(float(x.span.text))
        else:
                MetaScore.append(np.nan)
        G=sample.find('p', class_="sort-num_votes-visible")
        V_G=G.find_all('span',{"name":'nv'})
        Vot=V_G[0]["data-value"]
        Votes.append(Vot)
        if(len(V_G)>1):
                    Grs=(V_G[1]['data-value'])
                    Gross.append(Grs)
        else:
                    Gross.append(np.nan)
        y=sample.find('meta')
        Rating.append(float(y["content"]))
        Genre.append(sample.find('span', class_="genre").text[1:])
        print("Done")
```

```python
df=pd.DataFrame({"Name":Name, "Rating":Rating, "MetaScore":MetaScore, "Vote_Count":Votes, "Genre":Genre, "Gross":Gross, "Year":Year})
```
